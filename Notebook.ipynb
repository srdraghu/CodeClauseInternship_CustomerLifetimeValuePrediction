Data preprocessing and Modeling

```python
data.isnull().sum()
```

```python
data['Effective To Date']=pd.to_datetime(data['Effective To Date'])
data['Effective To Date'].to_timestamp
```

```python
lb=LabelEncoder()
data['Customer']=lb.fit_transform(data['Customer'])
data['State']=lb.fit_transform(data['State'])
data['Response']=lb.fit_transform(data['Response'])
data['Coverage']=lb.fit_transform(data['Coverage'])
data['Education']=lb.fit_transform(data['Education'])
data['EmploymentStatus']=lb.fit_transform(data['EmploymentStatus'])
data['Gender']=lb.fit_transform(data['Gender'])
data['Policy Type']=lb.fit_transform(data['Policy Type'])
data['Policy']=lb.fit_transform(data['Policy'])
data['Renew Offer Type']=lb.fit_transform(data['Renew Offer Type'])
data['Vehicle Class']=lb.fit_transform(data['Vehicle Class'])
data['Vehicle Size']=lb.fit_transform(data['Vehicle Size'])
data['Sales Channel']=lb.fit_transform(data['Sales Channel'])
data['Marital Status']=lb.fit_transform(data['Marital Status'])
data['Location Code']=lb.fit_transform(data['Location Code'])
```

```python
correlation_matrix = data.corr()
correlation_with_clv = correlation_matrix["Customer Lifetime Value"].sort_values(ascending=False)
print(correlation_with_clv)
```

```python
top_features = correlation_with_clv.index[1:11]
print("Top features correlated with SalePrice:\n", top_features)
```

```python
new_data = data[top_features]
new_data
```

```python
new_data['Customer Lifetime Value']=data['Customer Lifetime Value']
new_data
```

```python
print(new_data['Customer Lifetime Value'].skew())
sns.distplot(new_data['Customer Lifetime Value'])
```

```python
new_data['Customer Lifetime Value']=np.log(new_data['Customer Lifetime Value'])
sns.distplot(new_data)
print(new_data['Customer Lifetime Value'].skew())
```

```python
new_data['Total Claim Amount'].skew()
new_data['Total Claim Amount']=np.log(new_data['Total Claim Amount'])
```

```python
new_data.index.is_unique
```

```python
X=new_data.drop(columns='Customer Lifetime Value',axis=1)
Y=new_data['Customer Lifetime Value']

xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3,random_state=0)
print(xtrain.shape)
print(xtest.shape)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
print("Model :",model)

prediction = model.predict(xtest)
print("Predicted Values :",prediction)

acc = r2_score(ytest,prediction)*100
mae=mean_absolute_error(ytest,prediction)
mse=mean_squared_error(ytest,prediction)
```

```python
print(f"R2 score : {acc:.2f}")
print(f"Mean Absolute error {mae:.2f}")
print(f"Mean Squared error {mse:.2f}")
```
